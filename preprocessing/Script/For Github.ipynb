```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("ggplot")
```

```python
insert = {"name":["mike jamal","tays fahad","majed abdo","samy ramy","Mmar saad"],
     "Age":["25","32","12","41","85"],
     "Conttrry":["UK","EGY","UK","EGY","KSA"],
     "M/F":["M","F","M","F","M"],
     "Email":["AB@gmail.com","C@gmail.com","D@gmail.com","AP@gmail.com","A@gmail.com"],
     "buy":["Yes","No","no","Yes","No"]}

df = pd.DataFrame(insert)
df
```

```python
df.rename(columns={"name":"Name"}, inplace=True)
df
```

```python
df["First_Name"] = df["Name"].str.split(" ").str.get(0)
```

```python
df["Last_Name"] = df["Name"].str.split(" ").str.get(1)
```

```python
df
```

```python
df["First_Name"] = df["First_Name"].str.capitalize()
```

```python
df["Last_Name"] = df["Last_Name"].str.capitalize()
```

```python
df
```

```python
df.drop("Name", axis = 1, inplace=True)
```

```python
df
```

```python
df = df[["First_Name","Last_Name","M/F","Age","buy","Email","Conttrry"]]
```

```python
df
```

```python
df.rename(columns={"M/F":"Gender"},inplace=True)
```

```python
df
```

```python
df["Gender"] = df["Gender"].str.replace("M","Male").str.replace("F","Female")
```

```python
df
```

```python
df.rename(columns={"buy":"purchase"}, inplace=True)
```

```python
pd.get_dummies(df["purchase"])
```

```python
df["purchase"] = df["purchase"].str.capitalize()
```

```python
df
```

```python
pd.get_dummies(df["purchase"])
```

```python
pd.get_dummies(df["purchase"], drop_first=True).rename({"Yes":"purchase"}, axis= 1)
```

```python
df["purchase"] = pd.get_dummies(df["purchase"], drop_first=True).rename({"Yes":"purchase"}, axis= 1)
```

```python
df
```

```python
df["Email"] = df["Email"].str.lower()
```

```python
df
```

```python
df.rename({"Conttrry":"Country"},axis = 1, inplace = True)
df
```
