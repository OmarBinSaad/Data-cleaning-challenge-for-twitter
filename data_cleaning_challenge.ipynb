```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/data_cleaning_challenge.csv")
df
```

```python
df.info()
```

```python
df1 = df.drop(columns=["Unnamed: 9","Unnamed: 10"])
```

```python
df2 = df1[df1["Row Type"].notna()]
df2
```

```python
c = []
co= 0
for i in df2["Row Type"]:
    if "first name" in i:
     co += 1 
    c.append(co)
```

```python
c
```

```python
add = df2
add
```

```python
add["cv"] = c
```

```python
add
```

```python
names = add
```

```python
names = names[names["Row Type"].str.contains("first name")]
```

```python
names.drop(columns=["Speed1","Speed2","Electricity","Effort","Weight","Torque"], inplace=True)
```

```python
names = names[names["Row Type"] != "Row Type"]
```

```python
names
```

```python
names["Row Type"] = names["Row Type"].str[11:]
names["Iter Number"] = names["Iter Number"].str[11:]
names["Power1"] = names["Power1"].str[5:]
```

```python
names
```

```python
add
```

```python
names
```

```python
names = names.rename(columns={"Row Type":"first name","Iter Number":"last name","Power1":"date"})
names
```

```python
add = add[add["Row Type"] != "Row Type"]
add 
```

```python
add = add.dropna()
```

```python
Final = pd.merge(left=names , right=add, how="inner", on="cv")
Final
```

```python
Final = Final.drop(columns=["cv"])
Final
```

```python
Final.isna().any()
```
